# 學習使用TensorFlow提供的Pre-defined Model
* 本筆記參考[https://tf.wiki/en/basic/models.html](https://tf.wiki/en/basic/models.html)

````python
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
````

## 1.初始化預訓練的MobileNetV2的網路架構

````python
model = tf.keras.applications.MobileNetV2()
````

以下為宣告instance可挾帶的參數：
* **input_shape**:
    * input tensor 的 shape
    * 大多為 224 x 224 x 3
    * 大部分的模型，都有限制解析度的下限，如最低為 32 x 32 or 75 x 75
* **include_top**:
    * 是否fully-connected layer有包含在整個網路的最後
    * 預設為 True
* **weights**:
    * 預先訓練的權重(pre-trained)
    * 預設為 imagenet 
        * 表示在 ImageNet dataset 預先訓練的權重
    * 可以設為 None
        * 表示在所有權重的透過隨機來初始化所有的權重
* **classes**:
    * classes的數量
    * 預設為1000
    * 假如想修改include_top需設為True或weights設為None
* 細節可參考[https://keras.io/api/applications/](https://keras.io/api/applications/)

#### Set learning phase
* 有些預定義的模型，某些層在訓練階段與測試階段有不一樣的行為
* 因此，當在訓練此類模型時，你必須先手動設定模型的階段
```python
tf.keras.backend.set_learning_phase(True)
```

## 2. 使用MobileNetV2模型訓練花朵辨識(無pre-trained)
* **tf_flowers** 為花朵圖片的資料集
    * 共有5類花


[參考網站1](https://medium.com/@nutanbhogendrasharma/tensorflow-image-classification-with-tf-flowers-dataset-e36205deb8fc)
[參考網站2](https://tf.wiki/en/basic/models.html)

````python
# hyperparameter設定
num_epoch = 20
batch_size = 50
learning_rate = 0.001
````

````python
# 從 tfds 載入 tf_flowers 資料集，分割訓練資料70%與測試資料30%，並設定為監督式學習
(training_set, validation_set), dataset_info = tfds.load(
    'tf_flowers',
    split=['train[:70%]', 'train[70%:]'],
    with_info=True,
    as_supervised=True,
)
````

````python
dataset_info
````

````python
# 查看分類數
num_classes = dataset_info.features['label'].num_classes
num_classes
````

````python
# 查看資料數量
num_training_examples = 0
num_validation_examples = 0

for ex in training_set:
    num_training_examples += 1
for ex in validation_set:
    num_validation_examples += 1

print(f'Training Images數量: {num_training_examples}')
print(f'Validation Images數量: {num_validation_examples} \n')
````

### 資料集中的圖片並非皆相同大小

````python
# 抽取5張出來觀察
for i ,example in enumerate(training_set.take(5)):
    print(f"圖片 {i+1} 的shape: {example[0].shape} label: {example[1]}")
````

### 將資料集視覺化

````python
for image, label in training_set.take(1):
    break
image = image.numpy()
plt.figure() # 建立一個新的圖形
plt.imshow(image) # 顯示圖片，color map預設為RGB，(細節:https://www.cnblogs.com/denny402/p/5122594.html)
plt.title('Label {}'.format(label))
plt.grid(False)
plt.show()
````

### 資料預處理
#### tf.image.resize(圖片, (長,寬), method=...)
* method='nearest', 'gaussian',....[method細節](https://www.tensorflow.org/api_docs/python/tf/image/ResizeMethod)
#### .shuffle(SUFFLE_SIZE)
* 打亂資料集。但並不是把全部的資料做shuffle，假設SUFFLE_SIZE為N，則只會把前N個資料做shuffle
#### .batch(BATCH_SIZE)
* 將原本分散一筆一筆的資料，以批次的方式包起來，每個訓練的步驟中，就會都拿到同樣batch size的樣本進行訓練
#### .repeat()
* 重覆 dataset 的元素幾次，如果是count=2，那你可以做到在一個 epoch 內對每筆資料掃過兩次的效果，count=None時，即代表無限重覆


[shuffle, batch, repeat參考網站](https://ithelp.ithome.com.tw/m/articles/10268970)

````python
# 對資料進行預處理
# 1. 重新調整照片大小
# 2. 進行正規化
# 3. 對資料集做shuffle(打散，洗牌)
# 4. 設定batch size
t_data = training_set.map(lambda img, label: (tf.image.resize(img, (224,224)) / 255.0, label)).shuffle(1024).batch(batch_size)
v_data = validation_set.map(lambda img, label: (tf.image.resize(img, (224,224)) / 255.0, label)).batch(batch_size)
````

````python
# 初始化模型，將權重初始化為隨機，並設定分類為五類
model = tf.keras.applications.MobileNetV2(weights=None, classes=5)
````

````python
# 選擇Adam作為優化器
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
````

````python
for e in range(num_epoch):
    for images, labels in t_data:
        with tf.GradientTape() as tape:
            labels_pred = model(images, training=True)
            loss = tf.keras.losses.sparse_categorical_crossentropy(y_true=labels, y_pred=labels_pred)
            loss = tf.reduce_mean(loss)
        grads = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(grads_and_vars=zip(grads, model.trainable_variables))
    print(f"epoch{e} loss: {loss.numpy()}")
````

````python
# 驗證模型準確度
sparse_catgorical_accuracy = tf.keras.metrics.SparseCategoricalAccuracy()
for images, labels in v_data:
    y_pred = model(images, training=False)
    sparse_catgorical_accuracy.update_state(y_true=labels, y_pred=y_pred)
print(f"test accuracy: {sparse_catgorical_accuracy.result()}")
````

### 模型準確度:  24%

## 3. 使用MobileNetV2模型訓練花朵辨識(使用pre-trained model做transfer learning)
* 安裝 TensorFLow Hub
    * 裡面含有許多 pre-trained model 供我們做 transfer learning
* Transfer learning
    * a process where we take an existing trained model, and extend it to do additional work. This involves leaving the bulk of the model unchanged, while adding and retraining the final layers, in order to get a different set of possible outputs.(引用: [網站](https://medium.com/@nutanbhogendrasharma/tensorflow-image-classification-with-tf-flowers-dataset-e36205deb8fc))

### Transfer learning and fine-tuning
*　Pre-trained model是已經在大型數據集上訓練過的網路，通常是訓練在大規模的圖像分類上。我們透過這樣的模型，使用Transfer learning去客製化模型。
* 如果一個模型已經在夠大和普遍通用的資料集訓練過後，可以利用這個模型已學到的通用特徵，去繼續訓練，無須從頭開始

````python
import tensorflow_hub as hub
````

* 參數 **input_shape**:
    * input tensor 的 shape
    * 大多為 224 x 224 x 3
    * 大部分的模型，都有限制解析度的下限，如最低為 32 x 32 or 75 x 75

````python
URL = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
feature_extractor = hub.KerasLayer(URL, input_shape=(224,224,3))
````

### 將特徵提取層凍結
* 將  feature extractor layer 的變數都凍結，使待回訓練時只會更改到 final classifier layer
* 這是 transfer learning 的做法

````python
feature_extractor.trainable = False
````

### 藉由預訓練好的feature extractor layer製作模型

````python
model = tf.keras.Sequential([
        feature_extractor,
        tf.keras.layers.Dense(5)
])
model.summary()
````

### Compile 模型

````python
# 用compile函數定義loss function、optimizer及成效衡量指標(mertrics)
model.compile(
    # 使用adam作為優化函數
    optimizer = 'adam', 
    # 使用cross entropy作為分類的loss function
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    # 利用準確度來當作衡量標準
    metrics=['accuracy']
)
````

### 訓練模型 

````python
EPOCHS = 10
history = model.fit(t_data,
                   epochs=EPOCHS,
                   validation_data=v_data)
````

### 將訓練過程與驗證視覺化 

````python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(EPOCHS)

# 設定圖表中文字體
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1) # subplot(row, column, index)
plt.plot(epochs_range, acc, label='訓練準確度')
plt.plot(epochs_range, val_acc, label="驗證準確度")
plt.legend(loc='lower right') #加上圖例(右下角)
plt.title("訓練與驗證準確度")

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='訓練的 Loss')
plt.plot(epochs_range, val_loss, label='驗證的 Loss')
plt.legend(loc='upper right')
plt.title('訓練與驗證時的 Loss')
plt.show()
````

### 檢查預測 

````python
# 製作存分類標籤的陣列
class_names = np.array(dataset_info.features['label'].names)
print(class_names)
````

````python
image_batch, label_batch = next(iter(t_data)) # 取一批次的訓練資料
image_batch = image_batch.numpy()
label_batch = label_batch.numpy()
predicted_batch = model.predict(image_batch)
predicted_batch = tf.squeeze(predicted_batch).numpy() # 刪掉所有維度為1的維
predicted_indexs = np.argmax(predicted_batch, axis=-1) # 選出他們個別為哪一類(index)
predicted_class_names = class_names[predicted_indexs] # 將index對應到class name
print(predicted_class_names)
````

````python
print("真實的 labels: ", label_batch)
print("預測的 labels: ", predicted_indexs)
````

### 將預測結果視覺化
* plt.subplot()  [使用細節連結](https://steam.oxxostudio.tw/category/python/example/matplotlib-subplot.html)

````python
plt.figure(figsize=(10,10)) #長寬皆為10英寸
for i in range(30):
    plt.subplot(6,5,i+1) #在6*5的畫布上排在n+1的位置
    plt.subplots_adjust(hspace = 0.3) #設定子圖的間隔 hspace, wspace
    plt.imshow(image_batch[i])
    color = "blue" if predicted_indexs[i] == label_batch[i] else "red"
    plt.title(predicted_class_names[i], color=color)
    plt.axis("off") #關閉坐標軸
p = plt.suptitle("模型預測結果 (藍色: 正確, 紅色: 錯誤)", fontsize=24)
````

### 模型準確度:  90%
