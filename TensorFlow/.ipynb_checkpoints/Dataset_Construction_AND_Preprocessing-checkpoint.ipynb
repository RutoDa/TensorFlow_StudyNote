* TensorFlow 提供了module tf.data，裏頭包括了一套靈活的資料集建構API，可快速的建構資料輸入的管道，尤其適用於資料量巨大的情況下

## Dataset construction
* tf.data 的核心為類別 **tf.data.Dataset**
    * 裏頭包含了一系列的 iteratively accessible elements，每個elements包含一個或多個tensor
    * 比如說，對於一個由圖片組成的資料集，每個element可以是一個shape為(長,寬,channels)的圖片tensor，也可以是由圖片tensor和label組成的tuple

````python
import tensorflow as tf
import numpy as np
````

## Dataset preprocessing
* **tf.data.Dataset** 類別提供許多資料集預處理的方法
* 最常用的為:
    * **Dataset.map(f)**: 
        * 對資料集中的每個元素應用函數f，得到一個新的資料集
    * **Dataset.shuffle(buffer_size)**: 
        * 將資料集打亂(設定一個固定大小的Buffer，取出前 buffer_size 的element放入Buffer，並從Buffer中隨機取樣，取樣後的資料用後續資料替換）
        * ![](img/shuffle.png)
    * **Dataset.batch(batch_size)**:
        * 將資料集分成batch，也就次對每個 batch_size 個 element，使用 **tf.stack()** 在第0維合併成為一個 element
        * ![image.png](img/batch.png)
    * **Dataset.take(count)**
        * 建立一個資料集，並取原資料集count次的資料進資料集
    * 還有如 Dataset.repeat()（重複資料集的element）、Dataset.reduce()（與Map相對的聚合操作）等
        * ![image.png](img/repeat.png)
    * 詳細可[參考https://www.tensorflow.org/versions/r2.0/api_docs/python/tf/data/Dataset](https://www.tensorflow.org/versions/r2.0/api_docs/python/tf/data/Dataset) 或 [https://www.tensorflow.org/api_docs/python/tf/data/Dataset](https://www.tensorflow.org/api_docs/python/tf/data/Dataset)
    
* 圖片來自[動畫TensoFlow tf.data.dataset: Repeat, Batch, Shuffle - explained!](https://www.youtube.com/watch?v=c7G5W4Wv72Q)

#### 使用 Dataset.shuffle() 將資料打散後再設置批次，Buffer大小設置為10000
* **在機器學習中訓練模型需要將數據打亂**，這樣可以保證每批次訓練的時候所用到的數據集是不一樣的，可以提高模型訓練效果。
* shuffle的順序很重要，**應該先shuffle再batch**，如果先batch後shuffle的話，那麼此時就只是對batch進行shuffle，而batch裡面的資料順序依舊是有序的，那麼隨機程度會減弱(實際並未shuffle)。
* buffer_size=1:不打亂順序，既保持原序
* buffer_size越大，打亂程度越大
    * 一般最好的方式是buffer_size=Dataset_size


[以上引用(https://blog.csdn.net/qq_35608277/article/details/116333888)](https://blog.csdn.net/qq_35608277/article/details/116333888)

````python
# 設定圖表中文字體
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
````

````python
mnist_dataset = tf.data.Dataset.from_tensor_slices((train_data, train_label))
mnist_dataset = mnist_dataset.shuffle(buffer_size=10000).batch(4)

# 第一次
print("第一次取")
for images, labels in mnist_dataset:
    # images: [4, 28, 28, 1], labels: [4]
    fig, axs = plt.subplots(1,4) #1*4的畫布
    for i in range(4):
        axs[i].set_title(labels.numpy()[i])
        axs[i].imshow(images.numpy()[i, :, :, 0])
    plt.show()
    break #只抓第一批就好

# 第二次
print("第二次取")
for images, labels in mnist_dataset:
    # images: [4, 28, 28, 1], labels: [4]
    fig, axs = plt.subplots(1,4) #1*4的畫布
    for i in range(4):
        axs[i].set_title(labels.numpy()[i])
        axs[i].imshow(images.numpy()[i, :, :, 0])
    plt.show()
    break #只抓第一批就好
print("每一次都不一樣!")
````

#### 詳細的Dataset.shuffle()
引用[網站](https://tf.wiki/zh_hant/basic/tools.html)
* 設定一個固定大小為 buffer_size 的 Buffer
* 初始化時，取出資料集中的前 buffer_size 個元素放入　Buffer
* 每次需要從資料集中取元素時，即從buffer中隨機採樣一個元素並取出，然後從後續的元素中取出一個放回到之前被取出的位置，以維持buffer的大小
* 因此，Buffer的大小需要根據資料集的特性和資料排列順序特點來進行合理的設置。比如：
    * 當 buffer_size 設置為1 時，其實等價於沒有進行任何的資料打散
    * 當資料集的label順序分佈極為不均勻（例如二元分類時資料集前N個的label為0，後N個的label為1）時，較小的Buffer大小會使得訓練時取出的Batch資料很可能全為同一label，從而影響訓練效果
    * 一般而言，資料集的順序分佈若較為隨機，則Buffer的大小可較小，否則則需要設置較大的Buffer

## tf.data的<font color=red>平行化</font>策略
### Dataset.prefetch()
* 資料集的準備處理非常耗時，使得我們在每進行一次訓練前都需要花費大量的時間準備待訓練的資料，而此時GPU只能idle等待資料，造成了計算資源的浪費
    * ![](https://tf.wiki/_images/datasets_without_pipelining.png)
* tf.data的資料集對象為我們提供了 **Dataset.prefetch()** 方法，使得我們可以讓資料集對象 Dataset 在訓練時預取出若干個元素，使得在GPU 訓練的同時CPU 可以準備資料，從而提升訓練流程的效率
    * ![](https://tf.wiki/_images/datasets_with_pipelining.png)
* MNIST的實作
    ```python
    mnist_dataset = mnist_dataset.prefetch(buffer_size=tf.data.experimental.AUTOTUNE)
    ```
    * tf.data.experimental.AUTOTUNE 為由TensorFlow 自動選擇合適的數值
###  Dataset.map() 
* 假設用於訓練的電腦具有2核的CPU，我們想充分利用多核心的優勢對資料進行平行化變換
    ```python
    mnist_dataset = mnist_dataset.map(map_func=rot90, num_parallel_calls=2)
    ```
    * ![](https://tf.wiki/_images/datasets_parallel_map.png)
        * 這裡同樣可以將 num_parallel_calls 設置為 tf.data.experimental.AUTOTUNE 讓TensorFlow 自動選擇合適的數值


[引用網站(https://tf.wiki/zh_hant/basic/tools.html)](https://tf.wiki/zh_hant/basic/tools.html)

## Fetching elements from datasets
* 建立資料集並預處理完成後，我們想要疊代獲取資料集中的資料並用於訓練
* **tf.data.Dataset** 是一個Python的可疊代對象
    * 可用for迴圈取資料
    ```python
    for a, b, c, ... in dataset:
    ```
    * 也可用 **iter()** 創建一個python iterator，並用 next() 獲取下一個 element
    ```python
    it = iter(dataset)
    a_0, b_0, c_0, ... = next(it)
    a_1, b_1, c_1, ... = next(it)
    ```
    * Keras 支援使用 tf.data.Dataset 直接作為 input
        * 當使用 tf.keras.Model 的 fit() 和 evaluate() 方法時
        * 可以將參數中的input資料 x 指定為一個element格式為(輸入資料, 標籤資料)的 Dataset ，並忽略掉參數中的label資料 y
        ```python
        # 原本的方式
        model.fit(x=train_data, y=train_label, epochs=num_epochs, batch_size=batch_size)
        ```
        ``` python
        # 也可以直接傳入 Dataset
        # 由於已經通過 Dataset.batch() 方法劃分了資料集的batch
        # 所以不需要給定 batch_size
        model.fit(mnist_dataset, epochs=num_epochs)
        ```

## 實作: 貓與狗分類
* [資料集載點](https://www.kaggle.com/competitions/dogs-vs-cats/data?select=train.zip)

````python
import tensorflow as tf
import os
````

````python
# 設定 hyperparameter
num_epochs = 10
batch_size = 32
learning_rate = 0.001
# 設定資料集路徑
data_dir = 'D:/datasets/cats_vs_dogs'
train_cats_dir = data_dir + '/train/cats/'
train_dogs_dir = data_dir + '/train/dogs/'
test_cats_dir = data_dir + '/valid/cats/'
test_dogs_dir = data_dir + '/valid/dogs/'
````

````python
def decode_and_resize(filename, label):
    # 讀取原始檔案
    image_string = tf.io.read_file(filename)
    # decode JEPG 格式的圖片
    image_decoded = tf.image.decode_jpeg(image_string)
    # 重新調整大小，並標準化圖片
    image_resized = tf.image.resize(image_decoded, [256, 256]) / 255.0
    return image_resized, label
````

````python
# 建構訓練資料集
train_cat_filenames = tf.constant([train_cats_dir + filename for filename in os.listdir(train_cats_dir)])
train_dog_filenames = tf.constant([train_dogs_dir + filename for filename in os.listdir(train_dogs_dir)])
# 所有訓練資料的檔名路徑 (用concat合併tensor)
train_filenames = tf.concat([train_cat_filenames, train_dog_filenames], axis=0)
# 所有訓練資料的label(將 cat 類的標籤設為0，dog 類的標籤設為1)
train_labels = tf.concat([
        tf.zeros(train_cat_filenames.shape, dtype=tf.int32), 
        tf.ones(train_dog_filenames.shape, dtype=tf.int32)], 
        axis=0)
````

````python
# 建立資料集(用檔名)
train_dataset = tf.data.Dataset.from_tensor_slices((train_filenames, train_labels))
````

````python
# 透過map()來載入圖片資料到資料集
train_dataset = train_dataset.map(
    map_func=decode_and_resize, 
    num_parallel_calls=tf.data.experimental.AUTOTUNE
)
````

````python
train_dataset = train_dataset.shuffle(buffer_size=18750)    
train_dataset = train_dataset.batch(batch_size)
train_dataset = train_dataset.prefetch(tf.data.experimental.AUTOTUNE)
````

````python
# 建立一個簡單的CNN model
model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(32, 3, activation='relu', input_shape=(256, 256, 3)),
        tf.keras.layers.MaxPooling2D(),
        tf.keras.layers.Conv2D(32, 5, activation='relu'),
        tf.keras.layers.MaxPooling2D(),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(2, activation='softmax')
    ])
````

````python
# 用compile函數定義loss function、optimizer及成效衡量指標(mertrics)
model.compile(
    # 使用adam作為優化函數
    optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
    # 使用cross entropy作為分類的loss function
    loss=tf.keras.losses.sparse_categorical_crossentropy,
    # 利用準確度來當作衡量標準
    metrics=[tf.keras.metrics.sparse_categorical_accuracy]
)
````

````python
# 建構測試資料集
test_cat_filenames = tf.constant([test_cats_dir + filename for filename in os.listdir(test_cats_dir)])
test_dog_filenames = tf.constant([test_dogs_dir + filename for filename in os.listdir(test_dogs_dir)])
test_filenames = tf.concat([test_cat_filenames, test_dog_filenames], axis=-1)
test_labels = tf.concat([
    tf.zeros(test_cat_filenames.shape, dtype=tf.int32), 
    tf.ones(test_dog_filenames.shape, dtype=tf.int32)], 
    axis=-1)
````

````python
test_dataset = tf.data.Dataset.from_tensor_slices((test_filenames, test_labels))
test_dataset = test_dataset.map(decode_and_resize)
test_dataset = test_dataset.batch(batch_size)
````

````python
# 訓練模型
history = model.fit(train_dataset,
                    epochs=num_epochs,
                    validation_data=test_dataset)
````

````python
# 測試準確度
print(model.metrics_names)
print(model.evaluate(test_dataset))
````

### 將訓練過程與驗證視覺化

````python
acc = history.history['sparse_categorical_accuracy']
val_acc = history.history['val_sparse_categorical_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(num_epochs)

# 設定圖表中文字體
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(6, 6))
plt.subplot(1, 2, 1) # subplot(row, column, index)
plt.plot(epochs_range, acc, label='訓練準確度')
plt.plot(epochs_range, val_acc, label="驗證準確度")
plt.legend(loc='lower right') #加上圖例(右下角)
plt.title("訓練與驗證準確度")

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='訓練的 Loss')
plt.plot(epochs_range, val_loss, label='驗證的 Loss')
plt.legend(loc='upper right')
plt.title('訓練與驗證時的 Loss')
plt.show()
````

## TFRecord(資料集存儲格式)
* TFRecord 是TensorFlow 中的資料集存儲格式
* 當我們將資料集整理成TFRecord格式後，TensorFlow就可以高效地讀取和處理這些資料集，從而幫助我們更高效地進行大規模的模型訓練
* 可以理解為一系列序列化的 **tf.train.Example** 所組成的文件，而每一個 **tf.train.Example** 又由若干個 **tf.train.Feature 的字典**組成
```python
# dataset.tfrecords
[
    {   # example 1 (tf.train.Example)
        'feature_1': tf.train.Feature,
        ...
        'feature_k': tf.train.Feature
    },
    ...
    {   # example N (tf.train.Example)
        'feature_1': tf.train.Feature,
        ...
        'feature_k': tf.train.Feature
    }
]
```
* 為了將各種形式的資料集整理為TFRecord格式，我們可以對資料集中的每個element進行以下步驟：
    1. 讀取資料element到RAM內
    2. 將該element轉換為 **tf.train.Example** 物件
        *　由於每一個　**tf.train.Example**　由若干個 **tf.train.Feature** 的字典組成，因此需要先建立Feature的字典
    3. 將 **tf.train.Example** 物件序列化為 string，並透過預先定義的 **tf.io.TFRecordWriter** 寫入TFRecord 文件
* 而讀取TFRecord 資料則可按照以下步驟：
    1. 透過 **tf.data.TFRecordDataset** 讀入原始的TFRecord 文件，獲得一個 tf.data.Dataset 資料集物件
    2. 透過 Dataset.map 方法，對該資料集物件中的每一個序列化的 tf.train.Example 的 string 執行 tf.io.parse_single_example 函數，做反序列化

### 實作: 貓與狗分類的TFRecord 文件

````python
# 設定資料集路徑
data_dir = 'D:/datasets/cats_vs_dogs'
train_cats_dir = data_dir + '/train/cats/'
train_dogs_dir = data_dir + '/train/dogs/'
tfrecord_file = data_dir + '/train/train.tfrecords'

test_cats_dir = data_dir + '/valid/cats/'
test_dogs_dir = data_dir + '/valid/dogs/'
````

````python
# 建構訓練資料集
train_cat_filenames = [train_cats_dir + filename for filename in os.listdir(train_cats_dir)]
train_dog_filenames = [train_dogs_dir + filename for filename in os.listdir(train_dogs_dir)]
train_filenames = train_cat_filenames + train_dog_filenames
# 將 cat 類的標籤設為0，dog 類的標籤設為1
train_labels = [0] * len(train_cat_filenames) + [1] * len(train_dog_filenames)  
````

* 透過以下程式碼，疊代讀取每張圖片，並建立 **tf.train.Feature** 字典和 **tf.train.Example** 物件，序列化後寫入TFRecord檔

````python
with tf.io.TFRecordWriter(tfrecord_file) as writer:
    for filename, label in zip(train_filenames, train_labels):
         # 讀取資料集圖片到RAM
        image = open(filename, 'rb').read()    
        # 建立 tf.train.Feature 字典
        # 圖片是一個 Bytes物件
        # label是一個 Int物件
        feature = {                       
            'image': tf.train.Feature(bytes_list=tf.train.BytesList(value=[image])),  
            'label': tf.train.Feature(int64_list=tf.train.Int64List(value=[label]))   
        }
        # 透過字典建立 Example
        example = tf.train.Example(features=tf.train.Features(feature=feature)) 
        # 將Example序列化並寫入 TFRecord 文件
        writer.write(example.SerializeToString())   
````

#### tf.train.Feature 支持三種資料格式
1. tf.train.BytesList
    * string 或 binary files(如圖片)
2. tf.train.FloatList
3. tf.train.Int64List

### 讀取TFRecord 文件
* 以下程式碼為，讀取之前建立的 train.tfrecords 文件，並透過 Dataset.map 方法，使用 tf.io.parse_single_example 函數對資料集中的每一個序列化的 tf.train.Example 物件解碼

````python
# 讀取 TFRecord 文件
raw_dataset = tf.data.TFRecordDataset(tfrecord_file)   
````

````python
# 定義Feature結構，告訴decoder每個Feature的type是什麼
feature_description = { 
    'image': tf.io.FixedLenFeature([], tf.string),
    'label': tf.io.FixedLenFeature([], tf.int64),
}
````

````python
def parse_example(example_string): 
    """將 TFRecord 文件中的每一個序列化的 tf.train.Example decode"""
    feature_dict = tf.io.parse_single_example(example_string, feature_description)
    # decode JPEG圖片
    feature_dict['image'] = tf.io.decode_jpeg(feature_dict['image'])    
    return feature_dict['image'], feature_dict['label']
````

````python
dataset = raw_dataset.map(parse_example)
````
